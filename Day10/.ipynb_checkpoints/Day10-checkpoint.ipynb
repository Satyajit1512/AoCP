```python
grid = []

with open('./input.txt', 'r') as file:

     for line in file:
        
        print(line)
                                  
        if line != "\n":
                       
            line = [int(val) for val in line.rstrip('\n')]

            grid.append(line)
```

```python
import copy

def dfs(node, parent, visited, grad, target, start):
    
    if start == (0,2):
        
        print(node)
    
    m = len(grid)
    
    n = len(grid[0])
    
    x, y = node
    
    
    height = grid[x][y]
                    
    p_x, p_y = parent

    
    if parent == (-1,-1):

        grad = 1
    
    else:
        
        grad = height - grid[p_x][p_y]
                                
    
    if grad != 1:
                
        return
    
        
    if height == target:   
                    
        paths.append(tuple(visited))
        
        return 
    
        
    for dx,dy in [(-1,0),(0,-1),(0,1),(1,0)]:
                
        next_x = x + dx

        next_y = y + dy

        if (next_x) >=0 and (next_x <= m - 1) and (next_y >= 0) and (next_y <= n-1):
            
            if grid[next_x][next_y] - grid[x][y] == 1:

                next_coord = (next_x, next_y)

                if (next_coord) not in visited:

                    visited.add(next_coord)

                    dfs(next_coord, node, visited, grad, target, start)

                    visited.remove(next_coord)

    return 
```

```python
target = 9

paths = list()

for i in range(len(grid)):
    
    for j in range(len(grid[0])):
        
        if grid[i][j] == 0:
                                    
            visited = set()
            
            visited.add((i,j))
            
            dfs((i,j),(-1,-1), visited, 0, target, start = (i,j))
            
            visited.remove((i,j))
```
