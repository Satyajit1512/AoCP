```python
import re
import numpy as np
import functools
import sys
import math

line_no = 0

@functools.lru_cache(maxsize = 1024)
def dp(curr_coord, curr_cost):
            
    if curr_coord == target_coord:
                
        return curr_cost
    
    else:
        
        x, y = curr_coord
        
        if x + a['x'] <= target_coord[0] and y + a['y'] <= target_coord[1]:
                
            a_move = dp((x + a['x'], y + a['y']), curr_cost + 3) 
        
        else:
                        
            a_move = np.inf
        
        if x + b['x'] <= target_coord[0] and y + b['y'] <= target_coord[1]:
                                    
            b_move = dp((x + b['x'], y + b['y']), curr_cost + 1)
            
        else:
            
            b_move = np.inf
    
    
    return min(a_move, b_move)
  
    
#binary search in the grid
def search(a, b):
    
    p_x = target_coord[0]
    
    p_y = target_coord[1]
    
    a_x = list(a.values())[0]
    b_x = list(b.values())[0]
    a_y = list(a.values())[1]
    b_y = list(b.values())[1]
    
    
    n = a_y * b_x - a_x * b_y
    z = a_y * p_x - a_x * p_y
    b = z // n

    if z % n == 0 and (p_x - b * b_x) % a_x == 0:
        a = (p_x - b * b_x) // a_x
        return 3 * a + b

    return 0
        
#greedy approach
```

```python
final_ans = 0    
    
with open('./trial.txt', 'r') as file:

     for line in file:
                                          
        if line != "\n":
            
            if line_no % 3 == 0:
                
                a = {k:v  for k,v in zip(["x","y"],[int(val) for  val in re.findall(r"[0-9]+",line)])}
                                                    
            elif line_no % 3 == 1:
                
                b = {k:v  for k,v in zip(["x","y"],[int(val) for  val in re.findall(r"[0-9]+",line)])}
                                                
            else:
                
                target_coord = tuple([int(val) for  val in re.findall(r"[0-9]+",line)])
                            
                ans = dp((0,0), 0)
                
                dp.cache_clear()
                
                if ans != np.inf:
                
                    final_ans += ans
                    
            line_no += 1
```
