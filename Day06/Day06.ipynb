```python
from enum import Enum


class Direction(Enum):
    Up = 1
    Down = 2
    Left = 3
    Right = 4
    

    
def change_dir(direction:Direction):
    
    if direction == Direction.Up:
        return Direction.Right
        
    elif direction == Direction.Right:
        return Direction.Down
    
    elif direction == Direction.Down:
        return Direction.Left
    
    else:
        return Direction.Up

    

def check_legit(grid, rows , cols,next_x, next_y):
    
    if (next_x >= 0) and (next_x <= rows - 1) and (next_y >= 0) and (next_y <= cols-1):

        if grid[next_x][next_y] != "#":
            
            return True
    
    
    return False
    
    

def nights_watch(grid,start_pos,start_dir):
    
    rows = len(grid)
    cols = len(grid[0])
    
    exit_states = set()
    
    for row in (0,rows-1):
        for col in range(cols):
            exit_states.add((row,col))
    
    for col in (0,cols-1):
        for row in range(rows):
            exit_states.add((row,col))     
            
       
    curr = start_pos
    steps = 0
    curr_dir = start_dir
    
    visited = set()
    
    
    while curr not in exit_states:

        visited.add(curr)
                                                        
        found_next = False
                    
        while not(found_next):
                        
            if curr_dir == Direction.Up:
                                
                next_x = curr[0] - 1
                next_y = curr[1] 

                if check_legit(grid, rows , cols, next_x, next_y):
                    next_pos = (next_x, next_y)
                    found_next = True
             
            
            elif curr_dir == Direction.Down:
                
                next_x = curr[0] + 1
                next_y = curr[1] 

                if check_legit(grid, rows , cols, next_x, next_y):
                    next_pos = (next_x, next_y)
                    found_next = True
                    
            
            elif curr_dir == Direction.Right:
                
                next_x = curr[0]
                next_y = curr[1] + 1 

                if check_legit(grid, rows , cols, next_x, next_y):
                    next_pos = (next_x, next_y)
                    found_next = True
                    
            else:
                
                next_x = curr[0] 
                next_y = curr[1] - 1

                if check_legit(grid, rows , cols, next_x, next_y):
                    next_pos = (next_x, next_y)
                    found_next = True
                    
            
            if not(found_next):
                curr_dir = change_dir(curr_dir)


        curr = next_pos
    
    
    return len(visited) + 1 # + 1 for exit state
```

```python
def check_loop(i,j,grid,start_pos,start_dir,):
    
    rows = len(grid)
    cols = len(grid[0])
    
    exit_states = set()
    
    for row in (0,rows-1):
        for col in range(cols):
            exit_states.add((row,col))
    
    for col in (0,cols-1):
        for row in range(rows):
            exit_states.add((row,col))     
            
            
    if (i,j) in exit_states:
        
        exit_states.remove((i,j))
       
    
    curr = start_pos
    curr_dir = start_dir
    
    visited = set()
    
    
    while (curr,curr_dir) not in visited:
   
        visited.add((curr, curr_dir))
                                                        
        found_next = False
                    
        while not(found_next):
                        
            if curr_dir == Direction.Up:
                                
                next_x = curr[0] - 1
                next_y = curr[1] 

                if check_legit(grid, rows , cols, next_x, next_y):
                    next_pos = (next_x, next_y)
                    found_next = True
             
            
            elif curr_dir == Direction.Down:
                
                next_x = curr[0] + 1
                next_y = curr[1] 

                if check_legit(grid, rows , cols, next_x, next_y):
                    next_pos = (next_x, next_y)
                    found_next = True
                    
            
            elif curr_dir == Direction.Right:
                
                next_x = curr[0]
                next_y = curr[1] + 1 

                if check_legit(grid, rows , cols, next_x, next_y):
                    next_pos = (next_x, next_y)
                    found_next = True
                    
            else:
                
                next_x = curr[0] 
                next_y = curr[1] - 1

                if check_legit(grid, rows , cols, next_x, next_y):
                    next_pos = (next_x, next_y)
                    found_next = True
                    
            
            if not(found_next):
                curr_dir = change_dir(curr_dir)

        
        if next_pos in exit_states:
            
            return False
        
        curr = next_pos
        
    return True
```
