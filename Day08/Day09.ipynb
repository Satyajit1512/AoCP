```python
file_list = []

file_id = 0

with open('./input.txt', 'r') as file:

     for line in file:
            
        for i,val in enumerate(line.rstrip("\n")):
                        
            if i%2 == 0:
                
                file_list.append([file_id]*int(val))
                
                file_id += 1
                
            else:
                                                    
                file_list.append(['.']*int(val))
                
```

```python
for j in range(len(file_list)-1, -1, -1):
    
    if j % 1000 == 0:
        
        print(j)
            
    list_to_move = file_list[j]
        
    for i in range(0,j,1):
                
        candidate = file_list[i]
                
        index = 0
        
        while(index < len(candidate) and candidate[index] != '.'):
                        
            index += 1
        
        if len(candidate) - index >= len(list_to_move):
                        
            for last_index in range(len(list_to_move)-1, -1, -1):
                
                val = list_to_move[last_index]
                
                candidate[index] = val
                
                list_to_move[last_index] = '.'
                
                index += 1               
```
